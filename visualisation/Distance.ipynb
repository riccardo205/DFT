```python
import pandas as pd
import matplotlib.pyplot as plt

path_excel = "../dimer/distance2/distance2.csv"
# Load the CSV file. If your CSV has no header, specify header=None.
df = pd.read_csv(path_excel)

df.columns.tolist()
df
```

```python
import numpy as np
# Plot the third column (index 2) as x-axis and the fourth column (index 3) as y-axis.
distance = df['Interatomic Distance (Å)']
energy = df['Total Energy (eV)']
min_energy = np.min(energy)
stable_distance = distance[np.argmin(energy)]
plt.plot(distance[1:], energy[1:], marker='o', linestyle='-')
plt.xlabel("Interatomic Distance (Å)")
plt.ylabel("Total Energy (eV)")
plt.title("Carbon Dimer")
plt.ylim([-301, -292])

plt.text(1, -293, f"E={min_energy:.3f} eV \nD={stable_distance:.3f} Å", 
         bbox=dict(facecolor="white", edgecolor="black", boxstyle="round,pad=0.5"))


#plt.show()
plt.savefig("Relaxation_distance.png", dpi=300)
```
